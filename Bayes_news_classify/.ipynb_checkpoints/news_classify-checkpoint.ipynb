```python
import pandas as pd
import jieba
import numpy as np
```

```python
df_news = pd.read_table('./data/val.txt', names=['category','theme','URL','content'], encoding='utf-8')
df_news = df_news.dropna()
df_news.tail()
```

```python
df_news.shape
```

```python
content = df_news.content.values.tolist()
print(content[100])
```

```python
content_S = []
for line in content:
    current_segment = jieba.lcut(line)
    if len(current_segment) > 1 and current_segment != '\r\n':
        content_S.append(current_segment)
```

```python
content_S[1000]
```

```python
df_content = pd.DataFrame({'content_S': content_S})
df_content.head()
```

```python
stopwords = pd.read_csv('stopwords.txt', index_col=False, sep='\t', quoting=3,names=['stopword'],encoding='utf-8')
stopwords.tail(10)
```

```python
stopwords.stopword.values.tolist()
```

```python
def drop_stopwords(contents, stopwords):
    contents_clean = []
    all_words = []
    for line in contents:
        line_clean = []
        for word in line:
            if word in stopwords:
                continue
            line_clean.append(word)
            all_words.append(str(word))
        contents_clean.append(line_clean)
    return contents_clean,all_words
```

```python
contents = df_content.content_S.values.tolist()
stopwords = stopwords.stopword.values.tolist()
content_clean,all_words = drop_stopwords(contents,stopwords)
```

```python
df_content = pd.DataFrame({'contents_clean': content_clean})
df_content.head()
```

```python
df_all_words = pd.DataFrame({'all_words':all_words})
df_all_words.head()
```

```python
df_all_words.groupby(by=['all_words']).size()
```

```python
words_count = df_all_words.groupby(by=['all_words'])['all_words'].agg({'count': np.size})
words_count.head()
```

```python
words_count = words_count.reset_index().sort_values(by=['count'],ascending=False)
words_count.head()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10,5)

wordcloud = WordCloud(font_path='./data/simhei.ttf', background_color='white',max_font_size=80)
word_frequence = {x[0]:x[1] for x in words_count.head(100).values}
wordcloud = wordcloud.fit_words(word_frequence)
plt.imshow(wordcloud)
```

```python
df_train = pd.DataFrame({'contents_clean':content_clean, 'label':df_news['category']})
df_train.tail()
```

```python
df_train.label.unique()
```

```python
label_mapping = {"汽车": 1, "财经": 2, "科技": 3, "健康": 4, "体育":5, "教育": 6,"文化": 7,"军事": 8,"娱乐": 9,"时尚": 0}
df_train['label'] = df_train['label'].map(label_mapping)
df_train.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_train['contents_clean'].values, df_train['label'].values, random_state=1)
```

```python
x_train[0][1]
```

```python
words = []
for line_index in range(len(x_train)):
    try:
        words.append(' '.join(x_train[line_index]))
    except:
        print(line_index)
```

```python
words[0]
```

```python
len(words)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
texts = ["dog cat fish","dog cat cat","fish bird", 'bird']
cv = CountVectorizer()
cv_fit = cv.fit_transform(texts)
print(cv.get_feature_names())
print(cv_fit.toarray())
print(cv_fit.toarray().sum(axis=0))
```

```python
texts = ["dog cat fish","dog cat cat","fish bird", 'bird']
cv = CountVectorizer(ngram_range=(1,2))
cv_fit = cv.fit_transform(texts)
print(cv.get_feature_names())
print(cv_fit.toarray())
print(cv_fit.toarray().sum(axis=0))
```

```python
vec = CountVectorizer(analyzer='word', max_features=4000, lowercase=False)
vec.fit(words)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(vec.transform(words), y_train)
```

```python
test_words = []
for line_index in range(len(x_test)):
    try:
        test_words.append(' '.join(x_test[line_index]))
    except:
        print(line_index)
test_words[0]
```

```python
classifier.score(vec.transform(test_words), y_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(analyzer='word', max_features=4000, lowercase=False)
vectorizer.fit(words)
```

```python
classifier2 = MultinomialNB()
classifier2.fit(vectorizer.transform(words), y_train)
```

```python
classifier2.score(vectorizer.transform(test_words), y_test)
```

```python
vec2 = CountVectorizer(ngram_range=(1,4) ,analyzer='word', max_features=8000, lowercase=False)
vec2.fit(words)
classifier3 = MultinomialNB()
classifier3.fit(vec2.transform(words), y_train)
classifier3.score(vec2.transform(test_words), y_test)
```

```python
vectorizer2 = TfidfVectorizer(ngram_range=(1,4) ,analyzer='word', max_features=8000, lowercase=False)
vectorizer2.fit(words)
classifier4 = MultinomialNB()
classifier4.fit(vectorizer2.transform(words), y_train)
classifier4.score(vectorizer2.transform(test_words), y_test)
```
